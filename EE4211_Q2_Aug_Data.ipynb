```python
import requests
import json
import datetime
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
```

```python
def get_data(num_days, month, month_data):
    for day in range(0, num_days + 1):
        day_str = str(day)
        for hour in range(0, 24):
            hour_str = str(hour)
            for minute in range(3, 60, 20):
                minute_str = str(minute)
                site = f'https://api.data.gov.sg/v1/transport/carpark-availability?date_time={year}-{month.zfill(2)}-{day_str.zfill(2)}T{hour_str.zfill(2)}%3A{minute_str.zfill(2)}%3A{second_str.zfill(2)}'
                response_API = requests.get(site)
                data = response_API.text
                data = json.loads(data)
                if(not "items" in data):
                    continue
                timestamp = data["items"][0]["timestamp"]
                #print(site)
                print(timestamp)

                data = data["items"][0]["carpark_data"]
                file_title = "EE4211_data_" + month + ".json"
                with open(file_title, 'w') as fp:
                    json.dump(data, fp)
                dt = (pd.read_json(file_title))
                for heading in ("total_lots","lot_type","lots_available"):
                    dt[heading] = dt["carpark_info"].apply(lambda x: x[0][heading])
                dt = dt.drop(["carpark_info"], axis=1)
                month_data = month_data.append(dt)

# only to be run once
```

```python
year = "2022"
second_str = "59"
num_days = 31
month = "8"

aug_data = pd.DataFrame()

get_data(num_days, month, aug_data)

aug_data
```
