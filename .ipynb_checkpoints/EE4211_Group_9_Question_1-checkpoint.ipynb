We then add separate columns extracting hour and day from datetime to allow us to study data groups grouped by day and/or hour. A percentage is calculated for finding out the precentage of lots filled for every timestamp.

```python
import warnings
warnings.filterwarnings('ignore')
feb_data_loaded["update_datetime"] = pd.to_datetime(feb_data_loaded["update_datetime"])
feb_data_loaded["minute"] = feb_data_loaded["update_datetime"].dt.minute
feb_data_loaded["hour"] = feb_data_loaded["update_datetime"].dt.hour
feb_data_loaded["day"] = feb_data_loaded["update_datetime"].dt.day
feb_data_loaded["percent_lots_filled"] = (1 - (feb_data_loaded["lots_available"]/feb_data_loaded["total_lots"])) * 100
feb_data_loaded.head()
```

```python
feb_data_daily = feb_data_loaded.groupby("day", as_index = False)
for day in feb_data_daily.groups.keys():
    day_data = feb_data_daily.get_group(day)
    day_data_hourly = day_data.groupby("hour", as_index = False)
    print(day)
    print(day_data_hourly.groups.keys())
#     for hour in day_data_hourly.groups.keys():
#         print(hour)
#         hour_data = day_data_hourly.get_group(hour)
#         print(hour_data.groupby("minute").groups.keys())
```

As we can see above, there are no NaN values for lots available and we have each hour's data for everyday pf the month. So we have no missing data.

We first plot aggregate hourly data for the entire month.

```python
hourly_data = feb_data_loaded.groupby('hour', as_index = False)['percent_lots_filled'].mean()
hourly_data
```

```python
hourly_data.plot(y = "percent_lots_filled", xlabel = 'hour', ylabel = 'Lots filled (percentage)', xticks = hourly_data["hour"])
```

We can now plot hourly data for each day of the month

```python
for day in feb_data_daily.groups.keys():
    day_data = feb_data_daily.get_group(day)
    hourly_data_day = day_data.groupby('hour', as_index = False)['percent_lots_filled'].mean()
    title = "Day: " + str(day)
    hourly_data_day.plot(title = title, y = "percent_lots_filled", xlabel = 'hour', ylabel = 'Lots filled (percentage)', xticks = hourly_data_day["hour"])
```

From the above visualisations we can conclude that in general, parking availabilities are the maximum during working hours with a sligth drop in availability during lunch break. This makes sense as people will be going to malls/back home for getting or eating lunch. Parking lots are the most filled during the late hours of the evening (from around 6 pm  onwards) or the early hours of the morning.
